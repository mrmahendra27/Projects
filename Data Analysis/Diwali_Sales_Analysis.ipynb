# **Loading All Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import chardet

%matplotlib inline
warnings.filterwarnings('ignore')
```

# Checking the encoding format of file

```python
with open("../Files/diwali_sales.csv", 'rb') as file:
  result = chardet.detect(file.read())
```

```python
result
```

# Load the CSV file

```python
df = pd.read_csv('../Files/diwali_sales.csv', encoding=result['encoding'])
```

### Dropping Column with No Data

```python
df.drop(columns=["Status", "unnamed1"], inplace=True)
```

### Checking for null values in table

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df['Amount'].isnull()]
```

```python
df.shape
```

### Drop the row with null Amount column

```python
df.dropna(inplace=True)
```

```python
df.shape
```

### Change Data type of Amount column to int


```python
df['Amount'] = df['Amount'].astype('int')
```

### Updating Marital Status column values

```python
df['Marital_Status'] = df['Marital_Status'].apply(lambda x: "Married" if x == 1 else "Single")
```

# **Performing Exploratory Data Analysis**

```python
df.columns
```

# **Gender**

```python
plt.figure(figsize=(10, 5))
plt.title("No. of Buyers per Gender")
ax = sns.countplot(data=df, x="Gender")
plt.ylabel("Count")

for i in ax.containers:
  ax.bar_label(i)
```

```python
amount_by_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)

amount_by_gender
```

```python
plt.title("Amount spent by Gender")
sns.barplot(data=amount_by_gender, x="Gender", y="Amount")
plt.show()
```

```python
order_count_per_gender = df.groupby(['Gender'])['Orders'].sum()

order_count_per_gender
```

```python
plt.title("Orders Per Gender")
plt.pie(x=order_count_per_gender, labels=["Female", "Male"], autopct='%1.1f%%')
plt.show()
```

# **Marital Status**

```python
plt.figure(figsize=(10, 5))
plt.title("No. of Buyers per Marital Status")
ax = sns.countplot(data=df, x="Marital_Status")
plt.ylabel("Count")

for i in ax.containers:
  ax.bar_label(i)
```

```python
amount_by_marital_status = df.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)

amount_by_marital_status
```

```python
plt.figure(figsize=(8, 4))
plt.title("Sales by Marital Status")
sns.barplot(data=amount_by_marital_status, x="Marital_Status", y="Amount")
plt.xlabel("Marital Status")
plt.show()
```

```python
order_count_by_marital_status_and_state = df.groupby(['Marital_Status', 'State'])['Orders'].sum().reset_index()

order_count_by_marital_status_and_state
```

```python
plt.figure(figsize=(10, 5))
plt.title("Orders Count by Marital Status Per State")
sns.barplot(order_count_by_marital_status_and_state, x="Marital_Status", y="Orders", hue="State")
plt.legend(bbox_to_anchor=(1, 1, 0, 0))
plt.show()
```

```python
order_count_by_marital_status_and_gender = df.groupby(['Marital_Status', 'Gender'])['Orders'].sum().reset_index()

order_count_by_marital_status_and_gender
```

```python
plt.title("Orders Count by Marital Status Per Gender")
sns.barplot(order_count_by_marital_status_and_gender, x="Marital_Status", y="Orders", hue="Gender")
plt.show()
```
